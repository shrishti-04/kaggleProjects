# 5. PERFORM TEXT DATA CLEANING BY REMOVE PUNCTUATIONS

```python
# String module is super useful when dealing with text data
# String contains constants and classes for working with text
import string
string.punctuation
```

```python
text = '!I love# pandas., data analytics!!'
text_punc_remove = [char   for char in text if char not in string.punctuation   ]
text_punc_remove
```

```python
text_punc_join = ''.join(text_punc_remove)
text_punc_join
```

```python
# Let's define a function to remove punctuations
def remove_punc(message):
  text_punc_remove = [char  for char in message if char not in string.punctuation   ]
  text_punc_join = ''.join(text_punc_remove)
  return text_punc_join
```

```python
# Let's remove punctuations from our dataset 
alexa_df['reviews_punc_removed'] = alexa_df['verified_reviews'].apply(remove_punc)
```

# 6. PERFORM TEXT DATA CLEANING BY REMOVING STOPWORDS

```python
import nltk
from nltk.corpus import stopwords
import gensim
from gensim.utils import simple_preprocess


# import re
# from nltk.stem import PorterStemmer, WordNetLemmatizer
# from nltk.tokenize import word_tokenize, sent_tokenize
# from gensim.parsing.preprocessing import STOPWORDS
```

```python
# download stopwords
nltk.download('stopwords')
stopwords.words('english')
```

```python
# Add more stopwords by using extend
stop_words = stopwords.words('english')
stop_words.extend(['amazon', 'Amazon', 'alexa', 'Echo', 'Alexa', 'echo', 'device', 'dot', 'Dot', ])
```

```python
# Simple_preprocess converts a string into a series of lowered case tokens
# Let's try it on a sample dataset 
gensim.utils.simple_preprocess(alexa_df['verified_reviews'][0])
```

```python
# Remove stopwords and remove short words (less than 2 characters)
def preprocess(text):
  results = []
  for token in gensim.utils.simple_preprocess(text):
    if token not in stop_words and len(token) >= 1:
      results.append(token)

  return results
```

```python
# apply pre-processing to the text column
alexa_df['reviews_punc_stopwords_removed'] = alexa_df['reviews_punc_removed'].apply(preprocess)
```
